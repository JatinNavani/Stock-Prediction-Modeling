```python
from skopt import gp_minimize
from skopt.space import Real, Integer
from skopt.utils import use_named_args
from skopt.space import Categorical
import skopt
from skopt.space import Integer
```

```python
def calculate_mae(params):
    global aapl_data

    mean_window,std_window,sigma = params
    mean_window = mean_window * 5
    std_window = std_window * 5
    sigma = sigma * 0.25
    

    df = aapl_data.copy()
    if 'Date' in df.columns:
        df.set_index('Date', inplace=True)

    
    df['Log_Returns'] = np.log(df['Adj Close']/df['Previous Adj Close']) 


    df['Rolling_Mean'] = df['Log_Returns'].ewm(span=mean_window).mean()
    df['Rolling_Std'] = df['Log_Returns'].ewm(span=std_window).std()

    
    df['High_Prediction'] = ((df['Rolling_Mean'] + (sigma * df['Rolling_Std'])) * df['Adj Close']) +  df['Adj Close']
    df['Low_Prediction'] = ((df['Rolling_Mean'] - (sigma * df['Rolling_Std'])) * df['Adj Close'] ) +  df['Adj Close']

    # Shift Predictions down by 1 row
    df['High_Prediction'] = df['High_Prediction'].shift(1)
    df['Low_Prediction'] = df['Low_Prediction'].shift(1)

    # Calculate the Prediction as the mean of High_Prediction and Low_Prediction
    df['Prediction'] = df[['High_Prediction', 'Low_Prediction']].mean(axis=1)
    df.dropna(subset=['Prediction', 'Adj Close'], inplace=True)


    # Calculate Mean Absolute Error (MAE)
    mae = (df['Prediction'] - df['Adj Close']).abs().mean()

    return mae
```

```python
param_space = [
    skopt.space.Integer(1, 25, name='mean_window'),
    skopt.space.Integer(1, 25, name='std_window'),
    skopt.space.Integer(1, 12, name='sigma')
]
#mean_window=45, std_window=45, sigma=1
```

```python
@use_named_args(param_space)
def wrapped_objective_function(**params):
    
    param_values = [params['mean_window'], params['std_window'],params['sigma']]
    return calculate_mae(param_values)
```

```python
result_2 = gp_minimize(
    wrapped_objective_function, 
    dimensions=param_space, 
    n_calls=150,
    random_state=2310,
    noise=0.00001
)
```
